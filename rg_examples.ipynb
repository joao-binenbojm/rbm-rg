```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import Boltzmann as k

# Understanding the RG analysis of the 1D Ising model

# Coupling constants transform
def k_transform(k1,k2):
    k1_new = 0.25*np.log(np.cosh(2*k1 + k2)*np.cosh(2*k1-k2)) - 0.5*np.log(np.cosh(k2))
    k2_new = k2 + 0.5*np.log(np.cosh(2*k1+k2)/np.cosh(2*k1-k2))
    return (k1_new,k2_new)
# Defining params
itrs = 10
J = 10**-20
mu = 2*(10**-7)
B = 2*(10**-15)
T = 10
k1 = J/(k*T)
k2 = mu*B/(k*T)
params = np.zeros((2,itrs));
# Carrying out iteration across coupling space
for i in range(itrs):
    params[0,i] = k1
    params[1,i] = k2
    k1,k2 = k_transform(k1,k2)

# Plotting results in parameter space
fig = plt.figure()
plt.plot(params[0,:],params[1,:])
plt.xlabel('K1')
plt.ylabel('K2')
plt.title('Trajectory in coupling space')
plt.show()
```

```python
# Understanding the RG analysis of the 2D Ising model

# Coupling constants transform
def k_transform_2d(K,L,M):
    K_new = 0.25*np.log(np.cosh(4*K))
    L_new = 0.125*np.log(np.cosh(4*K))
    M_new = 0.125*np.log(np.cosh(4*K)) - 0.5*np.log(np.cosh(2*K))
    return (K_new, L_new, M_new)
# Defining params
itrs = 100
J = 10**(-22)
T = 50
params = np.zeros((3,itrs))
K = 30#J/(k*T)
L = 0
M = 0
print(K)
# Carrying out iteration across coupling space
t = list(range(itrs))
for i in t:
    params[0,i] = K
    params[1,i] = L
    params[2,i] = M
    K,L,M = k_transform_2d(K,L,M)

# Plotting results in parameter space
fig = plt.figure()
plt.plot(t, params[0,:])
plt.plot(t, params[1,:])
plt.plot(t, params[2,:])
plt.legend(['K','L','M'])
plt.xlabel('t')
plt.ylabel('Coupling Constants')
plt.title('Trajectory in coupling space')
plt.show()
```
