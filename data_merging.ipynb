```python
def store_many_disk(images, zip_name='merged.zip'):
    """ Stores an array of images to disk
        Parameters:
        ---------------
        images       images array to be stored
    """
    num_images = len(images)

    # Save all the images one by one (into a zip file)
    my_zip = ZipFile(zip_name, 'w')
    for i in tqdm(range(num_images)):
        # Processing step
        img = ( 255*(data[i,:,:] - data[i,:,:].min()) \
                     /(data[i,:,:].max() - data[i,:,:].min()) ).astype('uint8')
        Image.fromarray(img).convert('L').save(f"{i}.png")
        del img
        my_zip.write(f"{i}.png")
        os.remove(f"{i}.png")
```
